# Web Scraping NBA Data

```python
import requests
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
pd.options.display.max_columns = None
import time
```

## MVP Voting Data

```python
for year in range(1991, 2022):
    url_to_scrape = 'https://www.basketball-reference.com/awards/awards_{}.html'.format(year)
    page = requests.get(url_to_scrape)
    with open('data/{}.html'.format(year), "w+") as f:
        f.write(page.text)
```

```python
with open('data/1991.html') as f:
    page = f.read()
```

```python
soup = BeautifulSoup(page, "html.parser")
```

```python
soup.find('tr', class_="over_header").decompose()
```

```python
html_table = soup.find_all(id='mvp')
```

```python
mvp_1991 = pd.read_html(str(html_table))[0]
```

```python
mvp_1991['year'] = 1991
```

```python
mvp_1991.head()
```

```python
mvp_1991.shape
```

```python
for year in range(1992, 2022):
    with open('data/{}.html'.format(year)) as f:
        page = f.read()
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_="over_header").decompose()
    html_table = soup.find_all(id='mvp')
    df = pd.read_html(str(html_table))[0]
    df['year'] = year
    mvp_1991 = pd.concat([mvp_1991, df])
    
```

## Team Standings

```python
for year in range(1991, 2022):
    url_to_scrape = 'https://www.basketball-reference.com/leagues/NBA_{}_standings.html'.format(year)
    page = requests.get(url_to_scrape)
    with open('team-standings-by-year/{}.html'.format(year), "w+") as f:
        f.write(page.text)
```

```python
with open('team-standings-by-year/1991.html') as f:
    page = f.read()
```

```python
soup = BeautifulSoup(page, "html.parser")
```

```python
soup.find('tr', class_='thead').decompose()
```

```python
eastern = soup.find_all(id='divs_standings_E')
```

```python
eastern_df = pd.read_html(str(eastern))[0]
```

```python
eastern_df['year'] = 1991
eastern_df.rename(columns={'Eastern Conference': 'team'}, inplace=True)
eastern_df = eastern_df[ eastern_df['team'] != 'Central Division']
eastern_df
```

```python
western = soup.find_all(id='divs_standings_W')
western_df = pd.read_html(str(western))[0]
western_df.rename(columns={'Western Conference': 'team'}, inplace=True)
western_df['year'] = 1991
western_df = western_df[ (western_df['team'] != 'Midwest Division') & (western_df['team'] != 'Pacific Division')]
western_df
```

```python
combined_1991 = pd.concat([eastern_df, western_df])
```

```python
combined_1991
```

```python
for year in range(1992, 2022):
    with open('team-standings-by-year/{}.html'.format(year)) as f:
        page = f.read()
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_='thead').decompose()
    
    eastern = soup.find_all(id='divs_standings_E')
    eastern_df = pd.read_html(str(eastern))[0]
    eastern_df['year'] = year
    eastern_df.rename(columns={'Eastern Conference': 'team'}, inplace=True)
    eastern_df = eastern_df[ eastern_df['team'] != 'Central Division']
    
    western = soup.find_all(id='divs_standings_W')
    western_df = pd.read_html(str(western))[0]
    western_df.rename(columns={'Western Conference': 'team'}, inplace=True)
    western_df['year'] = year
    western_df = western_df[ (western_df['team'] != 'Midwest Division') & (western_df['team'] != 'Pacific Division')]

    combined = pd.concat([eastern_df, western_df])
    combined_1991 = pd.concat([combined_1991, combined])
```
