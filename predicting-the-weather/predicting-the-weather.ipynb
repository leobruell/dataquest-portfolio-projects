## Preparing the Data

### Dropping Unnecessary Columns

```python
cols_to_keep = ['PRCP','TMAX','TMIN','SNOW','SNWD']
```

```python
small_weather = weather[cols_to_keep].copy()
```

```python
small_weather.columns = ['precipitation','temp_max','temp_min', 'daily_snow','total_snow_depth']
```

```python
small_weather['precipitation'].mode()
```

```python
small_weather[['temp_max','temp_min']] = small_weather[['temp_max','temp_min']].fillna(value=small_weather.mean())
```

## Creating the Target Column

```python
small_weather['tomorrow_max'] = small_weather['temp_max'].shift(-1)
```

```python
small_weather.dropna(inplace=True)
```

```python
small_weather.head()
```

## First Machine Learning Model

We will start by using a ridge regression to predict tomorrow's max temperature. 

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X = small_weather.drop('tomorrow_max', axis=1)
y = small_weather['tomorrow_max']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25 , shuffle=False)
```

```python
ridge = Ridge()
ridge.fit(X_train, y_train)
predictions = ridge.predict(X_test)
rmse = mean_squared_error(y_test, predictions, squared=False)
rmse
```

```python
predictions
```

```python
actual_vs_predict = pd.DataFrame(y_test)
actual_vs_predict['predicted'] = predictions
```

```python
plt.plot(actual_vs_predict)
```

```python
def split_train_predict(df, model):
    X = df.drop('tomorrow_max', axis=1)
    y = df['tomorrow_max']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25 , shuffle=False)
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    actual_vs_predict = pd.DataFrame(y_test)
    actual_vs_predict['predicted'] = predictions
    print("RMSE: {}".format(mean_squared_error(y_test,predictions, squared=False)))
    return actual_vs_predict
```

```python
split_train_predict(small_weather, Ridge())
```

## Adding More Predictor Columns

```python
small_weather['30_day_avg_high'] = small_weather['temp_max'].rolling(window=30, closed='left').mean()
```

```python
small_weather['5_day_avg_high'] = small_weather['temp_max'].rolling(window=5, closed='left').mean()
```

```python
small_weather['min_temp_rolling_ratio'] = small_weather['temp_min'] / small_weather['30_day_avg_high']
small_weather['max_temp_rolling_ratio'] = small_weather['temp_max'] / small_weather['30_day_avg_high']
```

```python
small_weather.head(32)
```

```python
small_weather = small_weather.dropna()
```

```python
small_weather
```

```python
split_train_predict(small_weather, Ridge())
```

Above, we see that the RMSE value declined slightly when we added the new columns. 

## Adding average monthly temperature column

```python
small_weather["monthly_avg_high"] = small_weather["temp_max"].groupby(small_weather.index.month).apply(lambda x: x.expanding().mean())
small_weather["day_avg_high"] = small_weather["temp_max"].groupby(small_weather.index.day_of_year).apply(lambda x: x.expanding().mean())
```

```python
small_weather['monthly_avg_high'].plot()
```

```python
split_train_predict(small_weather, Ridge())
```

```python
split_train_predict(small_weather, Lasso())
```

## Conclusion

Above, we saw that adding in rolling, monthly, and daily averages resulted in a slight decrease in RMSE. Additionally, we see that Ridge regression is slightly more accurate than Lasso regression.
